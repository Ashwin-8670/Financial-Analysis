```python
# Calculating the daily percentage change in closing price
a['Percentage_Change'] = a['Close'].pct_change()
```

```python
a['Log returns'] = np.log(a['Close']/a['Close'].shift())
```

```python
#Daily standard deviation
a['Log returns'].std()
```

```python
#lifting to the square root of half and we have 252 working days in the market every year and square root of 0.5
volatility = a['Percentage_Change'].std()*252**.5
volatility
```

```python
str_vol = str(round(volatility, 3)*100)
print(str_vol,'%')
```

```python
fig, ax = plt.subplots()
a['Log returns'].hist(ax=ax, bins=50, alpha=0.6, color='b')
ax.set_xlabel("Log return")
ax.set_ylabel("Freq of log return")
ax.set_title("TSX volatility: " + str_vol + "%")
```

```python
# 5 Day moving average
a['MA10'] = a['Close'].rolling(5).mean()
a
```

```python
#exponetial moving average
a['EMA10'] = a['Close'].ewm(span=10, adjust=False).mean()
a
```

```python
#moving average is more reactive the ema
fig, ax = plt.subplots()
a[['MA10', 'EMA10']].loc['2020-12-01':].plot(ax=ax)
a['Close'].loc['2020-12-01':].plot(ax=ax, alpha=0.25)
```

```python
#moving average is more reactive the ema
fig, ax = plt.subplots()
a[['MA10', 'EMA10']].plot(ax=ax)
a['Close'].loc['2020-12-01':].plot(ax=ax, alpha=0.25)
```

```python
#moving average is more reactive the ema
fig, ax = plt.subplots()
a[['MA10', 'EMA10']]['2021-04-01':].plot(ax=ax)
a['Close'].loc['2021-04-01':].plot(ax=ax, alpha=0.25)
```

```python
exp1 = a['Close'].ewm(span=12, adjust=False).mean()
exp2 = a['Close'].ewm(span=26, adjust=False).mean()
a['MACD'] = exp1 - exp2
a['Signal line'] = a['MACD'].ewm(span=9, adjust=False).mean()
```

```python
fig, ax = plt.subplots()
a[['MACD', 'Signal line']].plot(ax=ax)
a['Close'].plot(ax=ax, alpha=0.25, secondary_y=True)
```

```python
high14 = a['High'].rolling(14).max()
low14 = a['Low'].rolling(14).min()
a['%K'] = (a['Close'] - low14)*100/(high14 - low14)
a['%D'] = a['%K'].rolling(3).mean()
```

```python
fig, ax = plt.subplots()
a[['%K', '%D']].loc['2020-11-01':].plot(ax=ax)
ax.axhline(80, c='r', alpha=0.3)
ax.axhline(20, c='r', alpha=0.3)
a['Close'].loc['2020-11-01':].plot(ax=ax, alpha=0.3, secondary_y=True)
```
